```python
# Data
df = sns.load_dataset('titanic')
```

```python
# Scatter plot

sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

```python
# scatter plot size of figrue size

plt.figure(figsize=(10,8))


sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

```python
# Scatter plot size

plt.figure(figsize=(10,8))
# change the scatter points colors

palette={'male':'green', 'female':'red'}

sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)

```

```python
# scatter plot
# scatter plot size of figure size

plt.figure(figsize =(8,4))

# change the scatter point colors

palette ={'male':'brown', 'female':'pink'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)

# labels

plt.xlabel('Age')
plt.ylabel('Fare')

plt.title("Scatter plot of Age vs Fare")

# saving a plot 300dpi , 600dpi png

plt.savefig('scatter_plot.png', dpi=500)

# saving a plot in vector file such as svg
plt.savefig('scatter_plot.svg')


```

```python
# style of a plot

sns.set_style('whitegrid')

# scatter plot
# scatter plot size of figure size

plt.figure(figsize =(8,4))

# change the scatter point colors

palette ={'male':'brown', 'female':'pink'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)

# labels

plt.xlabel('Age')
plt.ylabel('Fare')

plt.title("Scatter plot of Age vs Fare")
```

```python
sns.set_theme(style='darkgrid', font_scale=1.5, font='Times New Roman')

# scatter plot
# scatter plot size of figure size

plt.figure(figsize =(8,4))

# change the scatter point colors

palette ={'male':'brown', 'female':'pink'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)

# labels

plt.xlabel('Age')
plt.ylabel('Fare')

plt.title("Scatter plot of Age vs Fare")
```

## Box plot

```python
sns.set_theme(style='darkgrid', font_scale=1.5, font='Times New Roman')

# scatter plot
# scatter plot size of figure size

plt.figure(figsize =(8,4))

# change the scatter point colors

palette ={'First':'brown', 'Second':'pink', 'Third':'green'}
sns.boxplot(data=df, x='sex', y='age', hue='class', palette=palette)

# labels

plt.xlabel('Age')
plt.ylabel('Fare')

plt.title("Scatter plot of Age vs Fare")

# change legend position and title

plt.legend(loc='lower right', title='Class Name')
```

```python
sns.set_theme(style='darkgrid', font_scale=1.5, font='Times New Roman')

# scatter plot
# scatter plot size of figure size

plt.figure(figsize =(15,6))

# change the scatter point colors

palette ={'male':'brown', 'female':'green'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex',size='age', palette=palette)

# labels

plt.xlabel('Age')
plt.ylabel('Fare')

plt.title("Scatter plot of Age vs Fare")
```

```python
sns.set_theme(style='darkgrid', font_scale=1.5, font='Times New Roman')

# scatter plot
# scatter plot size of figure size

plt.figure(figsize =(15,6))

# change the scatter point colors

palette ={'male':'brown', 'female':'green'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex',size='fare', palette=palette)

# labels

plt.xlabel('Age')
plt.ylabel('Fare')

plt.title("Scatter plot of Age vs Fare")
```

## Boxen plot

```python
sns.set_theme(style='darkgrid', font_scale=1.5, font='Times New Roman')

# scatter plot
# scatter plot size of figure size

plt.figure(figsize =(8,4))

# change the scatter point colors

palette ={'First':'brown', 'Second':'pink', 'Third':'green'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=palette)

# labels

plt.xlabel('Genger')
plt.ylabel('Fare')

plt.title("Scatter plot of Age vs Fare")

# change legend position and title

plt.legend(loc='upper right', title='Class Name')
```

### Change the x tick lables

```python
sns.set_theme(style='darkgrid', font_scale=1.5, font='Times New Roman')

# scatter plot
# scatter plot size of figure size

plt.figure(figsize =(8,4))

# change the scatter point colors

palette ={'First':'brown', 'Second':'pink', 'Third':'green'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=palette)

# labels

plt.xlabel('Genger')
plt.ylabel('Fare')

plt.title("Scatter plot of Age vs Fare")

# change legend position and title

plt.legend(loc='upper right', title='Class Name')

# Change the x tick labels

plt.xticks(ticks=[0,1], labels=['MALE','FEMALE'])
```

### XTICKS ROTATION

```python
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=palette)

# labels

plt.xlabel('Genger')
plt.ylabel('Fare')

plt.title("Scatter plot of Age vs Fare")

# change legend position and title

plt.legend(loc='upper right', title='Class Name')

# Change the x tick labels

plt.xticks(ticks=[0,1], labels=['MALE','FEMALE'])

# x ticks rotation
plt.xticks(rotation = 45)
```

## Bar Plot using seaborn

```python
sns.barplot(data=df, x='sex', y='age', hue='class', palette=palette)

# labels

plt.xlabel('Genger')
plt.ylabel('Fare')

plt.title("Scatter plot of Gender vs Fare")

# change legend position and title

plt.legend(loc='upper right', title='Class Name')

# Change the x tick labels

plt.xticks(ticks=[0,1], labels=['MALE','FEMALE'])

# x ticks rotation
plt.xticks(rotation = 45)
```

### add error bar

```python
sns.barplot(data=df, x='sex', y='age', hue='class', palette=palette, capsize=0.1)

# labels

plt.xlabel('Genger')
plt.ylabel('Fare')

plt.title("Scatter plot of Gender vs Fare")

# change legend position and title

plt.legend(loc='upper right', title='Class Name')

# Change the x tick labels

plt.xticks(ticks=[0,1], labels=['MALE','FEMALE'])

# x ticks rotation
plt.xticks(rotation = 45)
```
