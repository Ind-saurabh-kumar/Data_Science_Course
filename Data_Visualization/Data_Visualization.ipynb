```python
# import libraries

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

### Load dataset

```python
# import titanic data from seaborn library

df = sns.load_dataset('titanic')
```

```python
df
```

### First graph plot


```python
df.plot(kind ='scatter', x='age',y='fare')
```

### change color using name and hashcode

```python
df.plot(kind='scatter',x='age',y='fare', color='red')
```

```python
df.plot(kind='scatter',x='age',y='fare', color='#6686')
```

### groupby

```python
df.head()
```

```python
df.groupby('sex')['fare'].mean()
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.ylabel('Average Fare')
plt.xlabel('Gender')
```

### Seaborn plots

```python
sns.barplot(data=df, x='survived', y='fare', color="green")
```

```python
sns.barplot(data=df, x='sex', y='fare', hue='class')
```

```python
sns.barplot(data=df, x='sex',y='age', hue='class')
```

```python
import plotly.graph_objects as go 
import pandas as pd

z_data = df

fig = go.Figure(data=[go.Surface(z=z_data.values)])
fig.update_layout(title="3D Graph", autosize=False, width=500, height=500, margin=dict(l=70, r=60, b=70, t=90))

fig.show()
```
