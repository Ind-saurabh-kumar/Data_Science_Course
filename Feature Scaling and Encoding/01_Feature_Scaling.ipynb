## 01. Min Max Scaling

```python
import pandas as pd 
from sklearn.preprocessing import MinMaxScaler

# Sample Data

data ={'Value':[10,20,30,40,50]}
df=pd.DataFrame(data)
df.head()

```

```python
# scale the data usign min max scaler
# Min-Max Scaling

scaler = MinMaxScaler()

df['Scaled_Value'] = scaler.fit_transform(df[['Value']])
print(df)
```

## 2. Standard Scalar or Z-score normalization

```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler


# Sample Data

data ={'Value':[10,20,30,40,50]}
df=pd.DataFrame(data)
df.head()
```

```python
# scale the data usign min max scaler
# Min-Max Scaling

scaler = StandardScaler()

df['Scaled_Value'] = scaler.fit_transform(df[['Value']])
print(df)
```

## 3. Robust Scalar

```python
from sklearn.preprocessing import RobustScaler

# Sample ata with outliers
data ={'Value':[10,20,30,40,50]}
df = pd.DataFrame(data)

# Robust Scaling

scaler = RobustScaler()
df['Scaled_Value'] = scaler.fit_transform(df[['Value']])
print(df)
```

## 4. Logrithmic scaling/normalization

```python
import numpy as np 
import pandas as pd 


# random data with outliers
data ={'Value':[100000,2000000,30000,400000,50000]}
df = pd.DataFrame(data)
df.head()
```

```python
# Log Transform

df['Value']= np.log(df['Value'])
df['Value_log2']= np.log2(df['Value'])
df['Value_log10']= np.log10(df['Value'])
df.head()
```

```python
import seaborn as sns
```

```python
sns.boxplot(df, x='Value')
```
